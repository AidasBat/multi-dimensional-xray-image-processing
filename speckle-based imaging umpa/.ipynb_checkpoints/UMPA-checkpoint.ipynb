Signal extraction in speckle-based imaging relies on a random (speckle) pattern generated from a sandpaper. The random intensity fluctiations mark the wavefront and get distorted when a sample is introduced in the beam.
The pattern distortions encode attenuation (speckle change in intensity), differential phase (speckle displacement) and dark-field (speckle blurring) signal information.

![title](SBI.png)

One can model the changes in the pattern as
$$
I^{\mathrm{model}}(\mathbf{r}) = T(\mathbf{r})\cdot\{\langle I_{0}(\mathbf{r}-\mathbf{u})\rangle +D(\mathbf{r}) \cdot [I_{0}(\mathbf{r}-\mathbf{u}) - \langle I_{0}(\mathbf{r}-\mathbf{u}) \rangle ] \} \, ,
$$

where: <br>
 $I_{0}(\textbf{r})$: Intensity of the reference pattern with no sample in the beam. <br>
 $I(\textbf{r})$: Pattern intensity with the sample. <br>
 $\textbf{r}$: Pixel position $\textbf{r}=(x,y)$. <br>
 $\textbf{u}(\textbf{r})$: Displacement of the speckle due to refraction in the sample $\textbf{u}=(u_{x}(\textbf{r}), u_{y}(\textbf{r}))$. <br>
 $T(\textbf{r})$: Transmission through the sample. <br>
 $\langle I (\textbf{r}) \rangle$: Mean intensity in the vicinity of $\textbf{r}$. <br>
 $D(\textbf{r}):$ Dark-field signal--reference pattern reduction in amplitude. <br>

Among others, [Unified Modulated Pattern Analysis (UMPA)](https://doi.org/10.1364/OE.474794) is an algorithm capable of retrieving the signal from the speckle data. A **stack of reference images** (sandpaper only) -- positioned at different positions perpendicular to the beam, and a **stack of sample images** (sandpaper and sample) -- at the same sandpaper positions, are measured.

In [UMPA](https://doi.org/10.1364/OE.474794), the signals are retrieved pixelwise (for each $\mathbf{r}$), the sample images are matched to the reference images considering $I^{\mathrm{model}}(\mathbf{r})$ with a least squares minimization approach in an a weighted (Hamming) analysis window $\Gamma$ of size $(2N+1)\times(2N+1)$.

$$
L(\mathbf{r})=\sum_{m=1}^{M} \sum_{w_{x}=-N}^{N} \sum_{w_{y}=-N}^{N} \Gamma(\mathbf{w}) \cdot \left| I_{m}^{\mathrm{model}}(\mathbf{r}-\mathbf{w}) - I_{m}(\mathbf{r}-\mathbf{w}) \right|^{2} \, ,
$$

where $L(\mathbf{r})$ is the cost function to minimize and $I_m$ indicates the sample image at the m-th sandpaper position.

Usually, we don't want to look at the differential phase separately but at the integrated phase shift:

```python
pc = phaseint_fourier(out['dx'], out['dy'])
```

```python
plt.figure(figsize=(5,5))
plt.title('integrated phase')
plt.imshow(pc)
plt.axis('off')
plt.colorbar(location='bottom')
plt.tight_layout()
```

Notice how there seems to be a noticeable ramp in the image which obscures the phase measured from the object. Let's try to correct it. Do you have ideas what could lead to such a ramp?

```python
mprof = pc.mean(axis=0)
x = np.arange(len(mprof))
p0 = np.polyfit(x, mprof, 1)
fit = np.polyval(p0, x)
plt.figure()
plt.plot(mprof, label='horizontal mean')
plt.plot(fit, label='fit')
plt.legend()
```

```python
ramp = np.ones_like(pc)*fit[None,:]
```

```python
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.title('integrated phase')
plt.imshow(pc)
plt.axis('off')
plt.colorbar(location='bottom')
plt.subplot(122)
plt.title('integrated phase - ramp')
plt.imshow(pc-ramp+ramp.mean(), vmin=-3, vmax=3)
plt.axis('off')
plt.colorbar(location='bottom')
plt.tight_layout()
```
