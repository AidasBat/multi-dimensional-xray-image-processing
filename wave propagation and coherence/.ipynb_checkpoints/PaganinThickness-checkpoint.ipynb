# Wave propagation and coherence

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.ndimage as ndi
import tifffile
import xraylib

plt.rcParams['image.cmap'] = 'gray'
plt.rcParams['image.interpolation'] = 'none'
plt.rcParams['figure.figsize'] = (10,6)
```

Define useful functions as we will need complex refractive numbers for different materials and energies.

```python
def get_n(name, E, density=None):
    """
    Get compound refractive index values: beta, delta, mu.

    Parameters
    ----------
    name : string
        Compound chemical composition or compound name from the list
        xraylib.GetCompoundDataNISTList().
    E : array
        Energy or list of energies in keV.
    density : float, optional
        Density of the compound in g/cm³. If the compound is from the
        NISTList it is not required. Check compound specifications with 
        xraylib.GetCompoundDataNISTByName(name)

    Returns
    ----------
    out : dictionary
        Values in dictionary entries "name", beta", "delta", "mu".
    """
    
    if density is None:
        try:
            density = xraylib.GetCompoundDataNISTByName(name)['density']
        except:
            print('Specify density, compound not in NIST database.')
            return

    def _n(name, E, density):
        beta  = xraylib.Refractive_Index_Im(name, E+.0, density)
        delta = 1 - xraylib.Refractive_Index_Re(name, E+.0, density)
        lam = xraylib.KEV2ANGST * 1e-10 / E  # in m
        mu = beta * 4 * np.pi / lam
        return dict(name=name, beta=beta, delta=delta, mu=mu)

    if np.ndim(E)==0:
        E = [E]
    compound = np.array([_n(name, e, density)['name'] for e in E]).squeeze()
    beta = np.array([_n(name, e, density)['beta'] for e in E]).squeeze()
    delta = np.array([_n(name, e, density)['delta'] for e in E]).squeeze()
    mu = np.array([_n(name, e, density)['mu'] for e in E]).squeeze()
    out = dict(name=compound, beta=beta, delta=delta, mu=mu)

    return out
```

## Thickness retrieval according to Paganin and Weitkamp

Remember that the complex wavenumber is
$$
n = (1-\delta) + i\beta ,
$$
and the wavefield after propagating a length z
$$
    \psi(z) = \psi_0 \cdot e^{iknz}
            =\psi_0 \cdot
                    \underbrace{e^{ik z}}_\text{vacuum} \cdot \underbrace{e^{-i\delta k z}}_\text{refraction} \cdot 
                    \underbrace{e^{-\beta k z}}_\text{attenuation}.
$$
and the measured intensity on a screen/detector
$$
I(z) = \lvert\psi(z)\rvert^{2}
$$

![title](Wavefront.png)

Choose parameters for your propagation calculation.

```python
E = 27 # in keV - beam energy
lam = 1.24e-9/E # in m - wavelength
z = 1e-2 # in m - sample to detector distance (i.e. propagation distance)
px = 1e-6 # in m - pixel size of detector
T = 2e-3 # in m - sample thickness

# According to the Weitkamp et al. paper (see pdf in folder, if interested):
print('Stay below the critical propagation distance: {:.2e} m.'.format((2*px)**2/lam))
```

Load the image "bone.tiff" and consider it an aperture. Are we in the near field or far field regime? Compute the Fresnel number `F` from aperture size `a`, wavelength `lam` and propagation distance `z`. Estimate the aperture size from the extent of the object. Use the formula given in the lecture.

```python
# load image
img = plt.imread('bone.tiff')
# calculate Fresnel number
a = 100*px
F = a**2 / (lam * z)
print(F)
```

Consider the sample as a flat object in air, 2 mm thick in the beam direction. Knowing the acquisition energy and choosing a material from `xraylib`, define the wavefront $\psi_{0}$ at the object's exit plane (i.e. after traversing its full thickness).
Look at the phase and amplitude components. Set $\psi_0 = 1$.

```python
material = 'Glass, Plate' #'Bone, Compact (ICRU)'
n = get_n(material, E) 
dn = (1-n['delta']) + n['beta']*1j

#mixed phase-amplitude
psi = 1 * np.exp(2j * np.pi / lam * dn * img * T  )
# pure phase
#psi = np.exp(2j * np.pi / lam * np.real(dn) * img * T  )

plt.figure()
plt.subplot(121)
plt.title('Wavefront phase')
plt.imshow(np.angle(psi), cmap='viridis')
plt.colorbar(orientation='horizontal')
plt.subplot(122)
plt.title('Wavefront amplitude')
plt.imshow(np.abs(psi), cmap='gray')
plt.colorbar(orientation='horizontal')
plt.tight_layout()
```

Assuming we had a detector directly at the surface of the sample, how would the image look? Plot a horizontal line through the image to assess the edges before propagating the wave further in the following steps.

```python
plt.figure()
plt.subplot(121)
plt.imshow(np.abs(psi)**2)
plt.axhline(250, c = 'y')
plt.colorbar(orientation = 'horizontal')
plt.subplot(122)
plt.plot((np.abs(psi)**2)[250,:])
plt.tight_layout()
```

This indicates that we are in the [Fresnel diffraction](https://en.wikipedia.org/wiki/Fresnel_diffraction) regime, and the Fresnel propagator can be used to check how the wavefield evolves in free-space. For $F \ll 1$ instead, i.e. the propagation distance being much larger than the aperture size, we use [Fraunhofer diffraction](https://en.wikipedia.org/wiki/Fraunhofer_diffraction).

Define the near field propagation function using the Fresnel kernel you learned in the lecture. Useful functions may include `np.fft.fftfreq()` and `np.meshgrid()`

```python
def nf_prop(a, z, lam, px):
    """\
    Near field free-space propagation of a over a distance z, with wavelength
    lam.
    """
    M, N = a.shape
    u = 2. * np.pi * np.fft.fftfreq(img.shape[1], d=px)
    v = 2. * np.pi * np.fft.fftfreq(img.shape[0], d=px)
    uu, vv = np.meshgrid(u, v, indexing='xy')
    u2 = uu**2 + vv**2
    kernel = np.exp(-1j * z  * lam * u2 / (4 * np.pi)) # 
    return np.fft.ifftn(np.fft.fftn(a) * kernel)
```

Now look at the wavefield components after propagation. What are we able to measure at the detector plane?

```python
img_propagated = nf_prop(psi, z+2e-02, lam, px=px)
hologram = np.abs(img_propagated)**2

plt.figure(figsize=(10,4.5))
plt.subplot(131)
plt.title('Wavefront phase')
plt.imshow(np.angle(img_propagated), cmap='viridis')
plt.colorbar(orientation='horizontal')
plt.subplot(132)
plt.title('Wavefront amplitude')
plt.imshow(np.abs(img_propagated))
plt.colorbar(orientation='horizontal')
plt.subplot(133)
plt.title('Hologram')
plt.imshow(hologram)
plt.colorbar(orientation='horizontal')
plt.tight_layout()
```

Plot a horizontal line through the sample in the hologram to see the effect of edge enhancement.

```python
plt.figure()
plt.subplot(121)
plt.imshow(hologram)
plt.axhline(250, c = 'y')
plt.colorbar(orientation = 'horizontal')
plt.subplot(122)
plt.plot(hologram[250,:])
plt.tight_layout()
```

We notice that fringes forming at the sample's interface. Having measured the hologram with a detector at a distance `z`, we can use the filter proposed by [Paganin et al.](https://doi.org/10.1046/j.1365-2818.2002.01010.x), described by [Weitkamp et al.](https://doi.org/10.1107/s0909049511002895), to retrieve the object thickness $T(x,y)$ and phase $\varphi(x,y)$ assuming the sample is homogenoeus and made of a single material. 
$$
T(x,y) = -\frac{1}{\mu} \mathrm{ln}\left(\mathcal{F}^{-1}\bigg\{ \frac{\mathcal{F}[I(x,y)]}{1+z\delta\mu^{-1}(u^{2}+v^{2})} \bigg\} \right) \, ,
$$

For now, we will concentrate on retrieving the thickness.

```python
def thickness(a, mu, delta, z, px):
    '''
    Retrieve object thickness from Weitkamp et al. paper.
    
    Parameters
    ----------
    a : 2D-array
        Input image.
    mu : float
        Linear attenuation coefficient.
    delta : float
        Real part of complex refractive index.
    z : float
        Distance between the object and the detector.
    px : float
        Pixel size.
    
    Returns
    ----------
    out : 2D-array
        Returns estimated object thickness in the beam direction.
    '''
    sh = a.shape
    u = 2 * np.pi * np.fft.fftfreq(sh[0], d=px)
    v = 2 * np.pi * np.fft.fftfreq(sh[1], d=px)
    uu, vv = np.meshgrid(u,v, indexing='ij')
    uv2 = uu**2 + vv**2
    
    num = np.fft.fftn(a)
    denom = 1 + ( z * delta / mu ) * uv2 
    
    return -1. / mu * np.log( np.real( np.fft.ifftn( num / denom ) ) ) 
```

```python
img_thickness = thickness(hologram, delta=n['delta'], mu=n['mu'], z=z, px=px)
```

```python
slc = img.shape[0]//2
```

```python
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.title('Simulated thickness {:,.2g}'.format(T*1000)+r' $\mu$m')
plt.imshow(img_thickness)
plt.axhline(slc, c='y')
plt.subplot(122)
plt.plot(img_thickness[slc])
plt.axhline(T, c='r', ls='--')
plt.tight_layout()
```

## Coherence

These simulated examples were made in ideal conditions. Test what happens for a polychromatic spectrum, i.e. if we **lose temporal coherence**. 
One way to generate spectra with different filtration, among others, is [SpekPy Web](https://spekpy.smile.ki.se). We take two spectra generated from a W anode, for 50 keV potential and 100 mAs: one filtered with 0.05mm of Ag and the second with 0.7mm of Al. You can save the output and read it using `pandas` python library (you might need to install it). If you'd like to try different energies and filters, feel free to try!

```python
data = pd.read_csv('spek_50keV_0.05mmAg.csv', skiprows=18, delimiter='\t', names=['Energy', '0.05mm Ag'] )
data1 = pd.read_csv('spek_50keV_0.7mmAl.csv', skiprows=18, delimiter='\t', names=['Energy', '0.7mm Al'])
data['0.7mm Al'] = data1['0.7mm Al'] # Append a column to the previous DataFrame
```

```python
data.head()
```

The `DataFrame` contains the spectrum Energy [keV] in the first column and the two filtered spectra Fluence [cm⁻² keV⁻¹]. Now, plot the fluence for both spectra and calculate (and plot) the mean energies of each.

```python
data.plot(x=0, y=[1,2])
plt.title(r'Fluence [cm$^{-2}$ keV$^{-1}$]')
E_Al = np.sum(data['0.7mm Al']*data['Energy'])/np.sum(data['0.7mm Al'])
E_Ag = np.sum(data['0.05mm Ag']*data['Energy'])/np.sum(data['0.05mm Ag'])
plt.axvline(E_Al, c='r', ls='--', label='Al: $\overline{E}$'+' = {:.2f} keV'.format(E_Al))
plt.axvline(E_Ag, c='k', ls='--', label='Ag: $\overline{E}$'+' = {:.2f} keV'.format(E_Ag))
plt.legend()
plt.ylim(0,None)
plt.xlim(0,50)
plt.show()
```

Add a column with the refractive index difference `dn` for each energy bin. Calculate $\psi_{0}$ and the holograms measured for each energy bin and add those columns.

```python
data['dn'] = [ (1-get_n(material, e)['delta']) + get_n(material, e)['beta']*1j for e in data['Energy']]
```

```python
data['Psi'] = [ np.exp(2j * np.pi * img * T * n / (1.24e-9/e) ) for e, n in zip(data['Energy'], data['dn'])]
```

```python
data['Holo'] = [ np.abs(nf_prop(p, z, 1.24e-9/e, px=px))**2 for e, p in zip(data['Energy'], data['Psi'])]
```

Define a slice to see how the final image varies when changing the DataFrame/energy range. Estimate measured holograms for the two spectra.

```python
sl = np.s_[:] 
```

```python
holo_Al = np.sum(data['Holo'][sl]*data['0.7mm Al'][sl]) / np.sum(data['0.7mm Al'][sl])
```

```python
holo_Ag = np.sum(data['Holo'][sl]*data['0.05mm Ag'][sl]) / np.sum(data['0.05mm Ag'][sl])
```

```python
vmin = 0.5
vmax = 1.2
plt.figure()
ax1 = plt.subplot(131)
plt.title('Image with Ag[0.05mm] spectrum\n'+'{:.2g} keV'.format(E_Ag))
im1 = ax1.imshow(holo_Ag, vmin = vmin, vmax = vmax)
ax1.axis('off')
plt.colorbar(im1, shrink=0.25)
ax2 = plt.subplot(132, sharex = ax1, sharey = ax1)
plt.title('Image with Al[0.7mm] spectrum\n'+'{:.2g} keV'.format(E_Al))
im2 = ax2.imshow(holo_Al, vmin = vmin, vmax = vmax)
plt.colorbar(im2, shrink=0.25)
ax2.axis('off')
ax3 = plt.subplot(133, sharex = ax1, sharey = ax1)
plt.title('Image monochromatic\n'+'{:.2g} keV'.format(E))
im3 = plt.imshow(hologram, vmin = vmin, vmax = vmax)
plt.colorbar(im3, shrink=0.25, ax = ax3)
plt.axis('off')
plt.tight_layout()
plt.show()
```

How does it affect thickness estimate? Remember that $\delta$ and $\mu$ depend on the energy. You can either first calculate the mean energy and calculate the quantities from that or calculate the mean value of the quantity using the energy range.

```python
E_Al = np.sum(data['0.7mm Al'][sl]*data['Energy'][sl])/np.sum(data['0.7mm Al'][sl])
E_Ag = np.sum(data['0.05mm Ag'][sl]*data['Energy'][sl])/np.sum(data['0.05mm Ag'][sl])
T_Al = thickness(holo_Al, delta=get_n(material, E_Al)['delta'], mu=get_n(material, E_Al)['mu'], z=z, px=px)
T_Ag = thickness(holo_Ag, delta=get_n(material, E_Ag)['delta'], mu=get_n(material, E_Ag)['mu'], z=z, px=px)
```

## Fraunhofer propagation

As mentioned before, in far-field propagation the assumptions of the near-field propagation do not hold anymore. The measured intensity in the far away measurement plane is now proportional to the Fourier transform of the wave function in the sample plane.
$$
I(u,v) \propto \lvert \mathfrak{F}\{\text{G}(x_0, y_0, z = 0)\}\rvert ^2
$$
Take the image `bone.tiff` and propagate it in the far field. First, we generate a phase-field in the bone:
* use the wavefront `wavefront.txt` as phase to propagate (choose a region of interest to match the shape of the array to the 'bone' image)
* use the 'bone' image as magnitude
* you may need `np.fft.fftshift` here

```python
img = plt.imread('bone.tiff')
y = img.shape[0]
x = img.shape[1]
```

```python
wf = np.loadtxt('wavefront.txt')
wf_crop = wf[:y,:x]

field = img * np.exp(1.j * wf_crop*img) 
# homogeneous phase
#field = img * np.exp(1.j * np.pi * img)

plt.figure()
plt.subplot(211)
plt.imshow(wf_crop, cmap = 'viridis')
plt.colorbar()
plt.subplot(212)
plt.imshow(np.angle(field), cmap = 'viridis')
plt.colorbar()
```

Next, show the intensity image on the screen in the measurement plane.

```python
#field = img * np.exp(1.j * wf_crop) 
field_prop = np.fft.fftshift(np.fft.fft2(field))
intensity = np.abs(field_prop)**2

plt.figure()
plt.imshow(intensity)
```
